```python
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import os
```

### Simple Script to turn barcodes into numpy array

```python
for filename in glob('./height_barcodes/*barcode_0*'):
    dim0 = np.loadtxt(filename, dtype=np.float32)
    dim0[dim0 == -1] = 1000.0
    assert(dim0.shape[1] == 2)
    assert(dim0.min() >= 0)
    print(filename, dim0.shape)
    np.save('./sublevelset_barcode_numpy/' + filename.split('/')[-1].replace('barcode_0', 'dim0_sublevel').replace('.txt', '.npy'), dim0)
```
